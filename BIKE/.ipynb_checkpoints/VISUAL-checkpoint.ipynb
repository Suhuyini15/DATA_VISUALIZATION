```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./bike_sales_data.csv')
df
```

## COUNTPLOT

```python
fig,axes = plt.subplots(1, 2, figsize=(12, 6))
sns.countplot(data=df,x='Age_Group' ,hue='Color',ax=axes[0])
axes[0].set_title('Countplot of Age group amd Color')

sns.countplot(data=df,x='Age_Group',hue='Order_Quantity',ax=axes[1],color='red')
axes[1].set_title('Coutnplot of Age group and Order Quantity')
plt.show()
```

```python
fig,axes = plt.subplots(1, 2, figsize=(12, 6))
sns.countplot(data=df,x='Age_Group',ax=axes[0])
axes[0].set_title('Countplot of Age group amd Color')

sns.countplot(data=df,x='Customer_Gender',ax=axes[1],color='red')
axes[1].set_title('Coutnplot of Age group and Order Quantity')
plt.show()
```

```python
fig,axes = plt.subplots(1, 2, figsize=(20, 12))
sns.countplot(data=df,x='State' ,hue='Color',ax=axes[0])
axes[0].set_title('Countplot of Age group amd Color')

sns.countplot(data=df,x='State',hue='Order_Quantity',ax=axes[1],color='red')
axes[1].set_title('Coutnplot of Age group and Order Quantity')
plt.show()
```

## HEATMAP

```python
correlation = df[['Size','Order_Quantity', ' Unit_Cost ', ' Unit_Price ', ' Profit ',
       ' Cost ', 'Revenue',]].corr()
```

```python
sns.heatmap(correlation,annot=True, square=True, fmt=".2f")
plt.show()
```
