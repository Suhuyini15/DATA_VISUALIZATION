```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./shopping_trends_updated.csv')
df.head(30)
```

```python
df = df.drop_duplicates()
```

## COUNTPLOT

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 6))
sns.countplot(x='Gender', data=df,ax=axes[0],hue='Size')
axes[0].set_title("Countplot of Gender")

sns.countplot(x='Category',data=df,ax=axes[1],hue='Gender')
axes[1].set_title("Countplot of Category")
plt.tight_layout()
plt.show()
```

## SCATTERPLOT

```python
df[['Age','Purchase Amount (USD)','Review Rating','Previous Purchases']].corr()
```

```python
plt.figure(figsize=(9,10))
sns.scatterplot(x='Previous Purchases',y='Purchase Amount (USD)',data=df,hue='Purchase Amount (USD)')
plt.show()
```

```python
plt.figure(figsize=(8,10))
sns.scatterplot(x='Age',y='Purchase Amount (USD)',data=df,color='green')
plt.show()
```

## PAIRPLOT

```python
sns.pairplot(data=df)
plt.show()
```

## HEATMAP

```python
correlation = df[['Age','Purchase Amount (USD)','Review Rating','Previous Purchases']].corr()
```

```python
sns.heatmap(correlation, annot=True, cmap='coolwarm', square=True, fmt=".2f")
plt.show()
```

## LINEPLOT

```python
Total = df['Season'].value_counts()

# Convert the Series to a DataFrame
count_df = Total.reset_index()
count_df.columns = ['Season', 'Count']
count_df = count_df.sample(frac=1, random_state=1).reset_index(drop=True)
count_df
```

```python
sns.lineplot(data=count_df,x='Season',y='Count')
plt.show()
```

## boxplot

```python
sns.boxplot(data=df, x='Gender',y='Previous Purchases')
plt.show()
```
